# Feature Engineering:

#  LabelEncoding : Converting Categorial features into Numerical type

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
data.salary = lbl_encoder.fit_transform(data.salary)
data.Department = lbl_encoder.fit_transform(data.Department)
```

```python
data.salary.value_counts()
```

```python
data.Department.value_counts()
```

```python
data.head()
```

```python
X = data.drop('left',axis =1)
Y = data.left
```

# Building a Random Forest Model

```python
from sklearn.model_selection import train_test_split
x_train , x_test ,y_train ,y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
std_scalar = StandardScaler()
x_train_scaled = std_scalar.fit_transform(x_train)
x_test_scaled = std_scalar.fit_transform(x_test)

x_train_scaled
```

```python
x_test_scaled
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()
random_forest.fit(x_train_scaled,y_train)
```

# Model Predictions

```python
pred = random_forest.predict(x_test_scaled)
pred
```

# Model Evaluation

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix = confusion_matrix(pred,y_test)
confusion_matrix
```

```python
accuracy = accuracy_score(pred,y_test)
print("Accuracy = " ,accuracy*100)
```

```python
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues', annot_kws={"color": 'red'})
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,pred)
report
```

# Feature Extraction

```python
scores_list = random_forest.feature_importances_
features = list(X.columns)
score_data = pd.DataFrame({"Feature":features,"Score":scores_list})
score_data.sort_values(by="Score",ascending =False)
```

# K-fold cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(random_forest,x_train_scaled,y_train,cv=5,scoring= "accuracy")
scores
```

```python
Avg_model_scoring =scores.mean()
```
